```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df=pd.read_csv(url, header=None,
               names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship',
                        'race','sex', 'capital-gain','capital-loss','hours-per-week', 'native-country','class'])
```

Create a sql db from adult dataset and name it sqladb

```python
sqladb=df
```

1. Select 10 records from the adult sqladb

```python
TenRec=sqldf("select * from sqladb LIMIt 10")
print(TenRec, end='')
```

2. Show me the average hours per week of all men who are working in private sector

```python
sql2="select workclass,avg([hours-per-week]) from sqladb where workclass like '%Private' group by workclass"
avghr=sqldf(sql2)
avghr
```

3. Show me the frequency table for education, occupation and relationship, separately

$Frequency Table For Education:$

```python
sql3a="select education, count(education) from sqladb group by education"
edu_freq=sqldf(sql3a)
edu_freq
```

$FrequencyTableForOccupation:$

```python
sql3b="select occupation, count(occupation) from sqladb group by occupation"
occ_freq=sqldf(sql3b)
occ_freq
```

$FrequencyTableForRelationship:$

```python
sql3c="select relationship, count(relationship) from sqladb group by relationship"
rel_freq=sqldf(sql3c)
rel_freq
```

4. Are there any people who are married, working in private sector and having a masters
degree

```python
#Check for different marital-status available
sql4a="select distinct [marital-status] from sqladb "
sqldf(sql4a)
```

```python
#ignore marital-status in Never-married,Divorced,Separated,Widowed
#check for relationship status after ignoring above marital-status
sql4b="select distinct [marital-status],relationship from sqladb where [marital-status] not in (' Never-married',' Divorced',' Separated',' Widowed')"
sqldf(sql4b)
```

```python
#ignore marital-status in Never-married,Divorced,Separated,Widowed
#check for relationship status after ignoring above marital-status
sql4b="select distinct [marital-status],relationship from sqladb where [marital-status] not in (' Never-married',' Divorced',' Separated',' Widowed')"
sqldf(sql4b)
```

```python
#ignore  relationship=Unmarried
sql4c="select distinct [marital-status],relationship from sqladb where [marital-status] not in (' Never-married',' Divorced',' Separated',' Widowed') and relationship<>' Unmarried'"
sqldf(sql4c)
```

```python
#find count of all records satisfying given criteria
sql4d="select count(*) from sqladb where [marital-status] not in (' Never-married',' Divorced',' Separated',' Widowed') and relationship<>' Unmarried' and education=' Masters' and workclass=' Private'"
sqldf(sql4d)

```

$CONCLUSION:$

There are 539 records where the people who are married, working in private sector and having a masters degree
